1. А точно нужно сортировать по вероятности? Я считаю нужно сортировать, т.к. весь смысл проделанной работы вычислить товар который больше всего заинтересует пользователя.

2. Какую метрику использовать? Я считаю что лучше всего качество предсказания показывает *Precision*, бизнесу интереснее будет видеть *Moneyprecision*

3. Сколько раз в неделю отпрпавляем рассылку? Если бмы будем проводить какие то акции, по видам товара которые интересовали пользователя, то стоит отправлять рассылку, можно ввести меру отклика, была ли покупка товара после проведенной рассылки, если пользователь уже давно, более месяца ничего не покупал, то рассылку необходимо проводить не чаще 1го раза в неделю, иначе если ему надоест захломление почты, он может просто отказаться от получения рассылки.

4. В какое время отправляем рассылку? Я считаю рассылку лучше проводить внерабочее время, вечером, в рабочее время или утром уведомление/письмо могут просто удалять не читая.

5. Будем отправлять одному юзеру много раз наши рекоммендации. Как добиться того, чтобы они хоть немного отличались? Можно вести таблицу рекомендованных товаров по пользователям, если после рекомендации товар/вид товара не купили, то отложить рекомендацию данного вида товара например на месяц.

6. Нужно ли, чтобы в одной рассылке были разные товары? Как определить, что товары разные? Как добиться того, чтобы они были разными? Товары необходимо рекомендловать по категоряим, например топовые товары каждой категории и товары этой же категории но средним ценником, с кратким перечнем самых важных хар-к, например ноутбук(частота проца, объем ОЗУ, объем ПЗУ, диагональ и вид матрицы). Если пользователь интересуется какой либо категорией товара то он увидит разброс цен сможет сделать быстрый анализ, и есть вероятность что перейдет по ссылке и купит его. Мне кажется не стоит рекомедовать товары из подкатегорий, ну тое зарядки итп.
